```python
import plotly.express as px

# Assuming 'Industry sector' and 'Subsector' are relevant columns in your DataFrame
# Industry and Subsector Comparison - Distribution of News

# Convert 'Industry sector' and 'Subsector' to strings, handling missing values
df['Industry sector'] = df['Industry sector'].astype(str)

df['Subsector'] = df['Subsector'].astype(str)

# Industry Distribution
import pandas as pd

# Assuming df is your DataFrame
df['Industry sector'] = df['Industry sector'].str.split(',')
df = df.explode('Industry sector').reset_index(drop=True)
df['Subsector'] = df['Subsector'].str.split(',')
df = df.explode('Subsector').reset_index(drop=True)

# Now df_expanded has each industry on a separate row

industry_distribution = df['Industry sector'].value_counts().reset_index()
industry_distribution.columns = ['Industry', 'Count']

# Subsector Distribution
subsector_distribution = df['Subsector'].value_counts().reset_index()
subsector_distribution.columns = ['Subsector', 'Count']

# Plot Industry Distribution
fig_industry = px.bar(industry_distribution, x='Industry', y='Count', title='Distribution of News Across Industries')
fig_industry.update_layout(xaxis_title='Industry', yaxis_title='Number of Articles')

# Plot Subsector Distribution
fig_subsector = px.bar(subsector_distribution, x='Subsector', y='Count', title='Distribution of News Across Subsectors')
fig_subsector.update_layout(xaxis_title='Subsector', yaxis_title='Number of Articles')

# Assuming 'Competitors' is the column containing competitors information
# Analyze Common Competitors
common_competitors = df.groupby('Industry sector')['Competitors'].apply(lambda x: ', '.join(set(str(x).split(','))).strip())
common_competitors = common_competitors.reset_index()
common_competitors.columns = ['Industry', 'Common Competitors']

# Plot Common Competitors
fig_competitors = px.bar(common_competitors, x='Industry', y='Common Competitors', title='Common Competitors Among Companies in the Same Industry')
fig_competitors.update_layout(xaxis_title='Industry', yaxis_title='Common Competitors')

# Show the plots
fig_industry.show()
fig_subsector.show()
fig_competitors.show()
```

```python
import pandas as pd
df = pd.read_excel('static/all_news.xlsx')

# Assuming df is your DataFrame with 'Company name' and 'Competitors' columns
# Note: You may need to clean and preprocess the 'Competitors' column based on your data
df['Industry sector'] = df['Industry sector'].str.split(', ')
df = df.explode('Industry sector').reset_index(drop=True)
df['Competitors'] = df['Competitors'].str.split(', ')
df = df.explode('Competitors').reset_index(drop = True)

# Drop rows where 'Competitors' column is either '0' or 'not mentioned'
df = df.drop(df[(df['Competitors'] == '0') | (df['Competitors'] == 'not mentioned') |(df['Competitors'] != 'Not specifically mentioned')].index)

df = df.dropna()

# Split the 'Competitors' column into lists of competitors
df['Competitors'] = df['Competitors'].str.split(', ')

# Create a new DataFrame with one row for each company-competitor pair
df_expanded = df.explode('Competitors').reset_index(drop=True)

# Check unique values in the 'Competitors' column after exploding
unique_values = df_expanded['Competitors'].unique()

# Identify non-string or missing values
non_string_values = [value for value in unique_values if not isinstance(value, str)]

# Handle non-string or missing values
df_expanded['Competitors'] = df_expanded['Competitors'].astype(str)

# Group by 'Company name' and create a set of unique competitors for each company
competitors_by_company = df_expanded.groupby('Company name')['Competitors'].apply(set).to_dict()

# Find common competitors among companies in the same industry
common_competitors = {}
for company, competitors in competitors_by_company.items():
    industry_companies = df_expanded[df_expanded['Company name'] == company]['Industry sector'].unique()
    for industry_company in industry_companies:
        if industry_company not in common_competitors:
            common_competitors[industry_company] = competitors
        else:
            common_competitors[industry_company] &= competitors

# Display the common competitors for each industry
for industry, common_comps in common_competitors.items():
    if industry != 'not mentioned' and ', '.join(common_comps) != 'Not specifically mentioned' and len(', '.join(common_comps)) != 0:
        print(f"Industry: {industry}")
        print(f"Common Competitors: {', '.join(common_comps)}")
        print()
```
